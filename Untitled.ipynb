```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200

t, u, v = np.loadtxt("./redress_ohmique.csv", dtype=float, delimiter=',', skiprows=2, unpack=True)
plt.plot(t, u)
plt.plot(t, v)
plt.show()

```

```python
# tension crete  crete

print(max(u)-max(v))
```

```python
def rase_periode(s):
    T = [] 
    compteur = 0

    for i in range(1, len(s), 5):
        if (s[i-5] - s[0]) * (s[i] - s[0]) <= 0:
            compteur += 1
            T.append(i)
    
    if compteur%2 == 1 :
        return T[-2]
    else :
        return T[-1]
    
print(rase_periode(u))
```

```python
Tu = rase_periode(u)
u = u[:Tu]
#t = t[:rase_periode(u)]

v = v[:Tu]
t = t[:Tu]

plt.plot(t, u)
plt.plot(t, v)
plt.show()
```

```python
def val_eff(s):
    #valeur efficace
    s = s[:rase_periode(s)]
    
    return np.sqrt(np.mean( s**2))


def val_eff_ctrd(s):
    #valeur efficace centre
    s = s[:rase_periode(s)]

    return np.sqrt( np.mean( s**2 ) - np.mean(s)**2 )

    
val_eff(u)


def moy(s):
    return np.mean(s)

#valeur moyenne, eff
print(moy(v))
print(val_eff(v))

```
