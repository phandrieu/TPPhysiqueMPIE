# Application au TP sur le redressement

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
f = 1
E = 10

def sinus(t):
    return np.sin( 2*np.pi*f * t) * E


def diode(t):
    e = sinus(t)
    return (1/2)*(e + np.abs(e))
```

```python

t = np.linspace(0, 5, 1000)
e = sinus(t)
s = diode(t)


plt.plot(t, e)
plt.plot(t, s)

plt.show()
```

```python
def moyenne_int(e):
    T = 1/f

    return quad(e, 0, T, args=())
```

```python
#moyennes
moyenne_int(sinus)
moyenne_int(diode)
```

```python
def carre(e):
    return lambda x: e(x) * e(x)

def variance(e):
    return moyenne_int(carre(e))[0] - (moyenne_int(e)[0])**2

variance(diode)
```

```python
def graetz(t):
    sig = sinus(t)
    return abs(sig)
```

```python
t = np.linspace(0, 5, 1000)
e = sinus(t)
s = graetz(t)


plt.plot(t, e)
plt.plot(t, s)

plt.show()
```
